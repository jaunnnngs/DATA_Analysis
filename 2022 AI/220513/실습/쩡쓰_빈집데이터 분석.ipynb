```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
#데이터셋 전처리
X = house[['연도', '도시지역면적', '비만율', '유치원 원아수', '초등학교 학생수', '토지거래면적',
       '혼인건수', '노인 천명당 노인여가복지시설수', '1인가구비율', '등록외국인 현황', '유아 천명당 보육시설수',
       '지가변동률', '출생아수', '학급당 학생수', '대학교수', '독거노인가구비율', '유치원수',
       '인구 십만명당 문화기반시설수', '인구 천명당 사설학원수', '주택수', '초등학교수', '평균연령', '빈집수']]
y = house['빈집지수']

sc = StandardScaler()
X = sc.fit_transform(X)

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.1 , random_state=0)
```

```python
print(len(x_train), len(x_test))
```

```python
#VIF로 다중공산성 확인해서 제거 할 변수 제거해보기.

Xs = house[['연도', '도시지역면적', '비만율', '유치원 원아수', '초등학교 학생수', '토지거래면적',
       '혼인건수', '노인 천명당 노인여가복지시설수', '1인가구비율', '등록외국인 현황', '유아 천명당 보육시설수',
       '지가변동률', '출생아수', '학급당 학생수', '대학교수', '독거노인가구비율', '유치원수',
       '인구 십만명당 문화기반시설수', '인구 천명당 사설학원수', '주택수', '초등학교수', '평균연령', '빈집수',
       '빈집지수']]
Xs = sm.add_constant(Xs)

vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(Xs.values, i) for i in range(Xs.shape[1])]
vif['predictor'] = Xs.columns
vif
# 5 이상 (유의) / 10 이상 (다중공선성 있다고 판단 - 변수 제거하여 사용)
```

```python
from sklearn.decomposition import PCA
# 주성분 분석
pca = PCA(n_components=23)
pca.fit(X)
```

```python
#기존 데이터 셋이 주성분으로는 어떻게 표현 되는지 확인
p_component = pca.transform(X)
pd.DataFrame(data = p_component).head()
```

```python
# 주성분이 기존변수들로 어떻게 표현되는지 확인
abs_house = abs(pd.DataFrame(pca.components_, columns=house.columns[1:-1]))
# 음양관계 상관없이 절댓값이 클수록 영향력 업!
```
